Projeto Python Básico
- Descrição:
Você foi contratado para desenvolver um sistema básico de gerenciamento de funcionários para uma pequena empresa. O sistema deve ser capaz de receber e armazenar informações sobre os funcionários e realizar operações básicas de manipulação desses dados.

- Requisitos:
Utilize a função input() realizar um cadastro coletando dados do usuário, como nome, cidade e estado em que reside, idade, escolaridade, cargo e salário.

- Funcionalidades do Sistema:
O sistema deve oferecer as seguintes funcionalidades: Adicionar um novo funcionário ao sistema, solicitando informações como nome, cargo, salário, e-mail, etc. Atualizar as informações de um funcionário existente com base em um identificador único (como um número de identificação ou nome). Exibir informações de um funcionário específico. Listar todos os funcionários cadastrados. Permitir a remoção de um funcionário do sistema.

- Observações Finais:
Certifique-se de implementar um sistema de identificação único para cada funcionário. Valide as entradas do usuário para garantir que as operações sejam realizadas corretamente e sem erros. Documente adequadamente seu código com comentários explicativos para ajudar na compreensão.

```python
funcionarios = [] # Lista para armazenar os funcionários
# Loop principal
while True:
  print("\nGerenciamento de Funcionários") # Exibição do menu principal
  print("==============================")
  opcao = input("\nDigite o número desejado: \n1 - Adicionar Funcionário \n2 - Atualizar informações do Funcionário \n3 - Listar os Funcionários Cadastrados \n4 - Remover cadastro do Funcionário \n5 - Sair \n")

  if opcao.isdigit() and int(opcao) in range(1, 6): # Verifica a entrada para não receber dados inválidos o método isdigit verifica se todos os caracteres são digitos
        opcao = int(opcao)
  else:
        print("Entrada inválida! Digite um número entre 1 e 5.")
        continue
  if opcao == 1:  # Opção 1: Adicionar um novo funcionário
    print("\nAdicionar Funcionário(a):")
    print("==============================")

    funcionario = {} # Dicionário para armazenar os dados do funcionário

    funcionario['id'] = len(funcionarios) + 1  # ID único baseado no tamanho da lista

    while True:
            nome = input("Digite o nome: ").title() # Inserção do nome do funcionário e colocando a primeira letra maiúscula
            if nome.replace(' ', '').isalpha(): # O replace remove os espaços vazios e o isalpha verifica se o valor digitado são letras e não está vazio
                funcionario['nome'] = nome
                break
            print("O campo 'Nome' não pode ser preenchido com números e não pode ficar vazio. Digite um valor válido!")

    while True:
            cidade = input("Digite a cidade: ").title() # Inserção da cidade do funcionário e coloca a primeira letra maiúscula
            if cidade.replace(" ", "").isalpha(): # O replace remove os espaços vazios e o isalpha verifica se o valor digitado são letras e não está vazio
                funcionario['cidade'] = cidade
                break
            print("O campo 'Cidade' não pode ser preenchido com números e não pode ficar vazio. Digite um valor válido! !")

    while True:
            estado = input("Digite o estado com duas letras (ex: São Paulo = SP): ").upper() # Inserção do estado e conversão para maiúsculas
            if len(estado) == 2 and estado.isalpha(): # O isalpha verifica se o valor digitado são letras, não está vazio e tem dois digitos
                funcionario['estado'] = estado
                break
            print("Estado inválido! Digite exatamente duas letras.")

    while True:
            idade = input("Digite a idade(apenas números): ") # Inserção da idade do funcionário
            if idade.isdigit() and int(idade) > 15 and int(idade) <= 75: # Verificando se o valor digitado são números e se está entre 16 e 75 anos
                funcionario['idade'] = int(idade)
                break
            print("Idade inválida! Digite uma idade válida, entre 16 e 75.")

    # Dicionário para mapear entradas de escolaridade para descrições completas
    escolaridade_map = {
        'f': 'Ensino Fundamental',
        'm': 'Ensino Médio',
        's': 'Ensino Superior'
    }

    while True:
            escolaridade = input("Digite a escolaridade:\nF - Fundamental \nM - Médio \nS - Superior \n").lower() # Inserção da escolaridade do funcionário, coloca para minuscula a letra digitada
            if escolaridade in escolaridade_map: # Verificando se a escolaridade é válida
                funcionario['escolaridade'] = escolaridade_map[escolaridade] # Inserção da descrição da escolaridade no dicionário
                break
            print('Escolaridade inválida! Digite: F, M ou S.')
    while True:
            cargo = input("Digite o cargo: ").title() # Inserção do cargo do funcionário e coloca a primeira letra para maiúscula
            if cargo.replace(" ", "").isalpha(): # O replace remove os espaços vazios e o isalpha verifica se o valor digitado são letras e não está vazio
                funcionario['cargo'] = cargo
                break
            print("O campo 'Cargo' não pode estar vazio!")
    while True:
            salario = input("Digite o salário: ") # Inserção do salário do funcionário
            if salario.isdigit() and int(salario) > 0: # Verificando se o valor digitado são números e maior que zero
                funcionario['salario'] = int(salario)
                break
            print("Salário inválido! Digite um valor maior que zero.")
    while True:
            email = input("Digite o e-mail: ") # Inserção do email
            if '@' in email and (email.endswith('.com') or email.endswith('.com.br')): # Validação do email se tem @, termina com .com ou .com.br
                funcionario['email'] = email
                break
            print("E-mail inválido!")

    funcionarios.append(funcionario) # Inserção do funcionário no Dicionário de Funcionarios
    print(f"Funcionário(a)'{funcionario['nome']}', adicionado com sucesso!") # Exibição do funcionário adicionado com sucesso
    print(funcionarios) # Exibição da lista de funcionários

  elif opcao == 2: # Opção 2: Atualizar as informações de um funcionário existente
    print("\nAtualizar Funcionário(a):")
    print("==============================")

    if not funcionarios:
        print("Não há funcionários cadastrados.")
    else:
      id_funcionario = int(input('Digite o id: '))  # Solicita o ID do funcionário a ser atualizado
      for funcionario in funcionarios:  # Itera sobre a lista de funcionários
         if funcionario['id'] == id_funcionario:  # Verifica se o ID existe na lista de funcionários
            print(f"Funcionário encontrado: {funcionario['nome']}")  # Exibe o nome do funcionário encontrado
            while True:
              print("\nO que deseja atualizar?")
              print("1 - Nome")
              print("2 - Cidade")
              print("3 - Estado")
              print("4 - Idade")
              print("5 - Escolaridade")
              print("6 - Cargo")
              print("7 - Salário")
              print("8 - E-mail")
              print("9 - Outro")
              print("10 - Finalizar atualização")
              opcao_atualizar = input("Escolha uma opção: ")
              if opcao_atualizar.isdigit() and int(opcao_atualizar) in range(1, 11): # Verifica a entrada para não receber dados inválidos o método isdigit verifica se todos os caracteres são digitos
                opcao_atualizar = int(opcao_atualizar)
              else:
                print("Opção inválida. Tente novamente.")
                continue

              if opcao_atualizar == 1:
                while True:
                   nome = input("Digite a atualização do nome: ").title() # Inserção do nome e coloca a primeira letra maiúscula
                   if nome.replace(" ", "").isalpha(): # O replace remove os espaços vazios e o isalpha verifica se o valor digitado são letras e não está vazio
                      funcionario['nome'] = nome
                      break
                   print("O campo 'Nome' não pode estar vazio!")

              elif opcao_atualizar == 2:
                while True:
                  cidade = input("Digite a cidade: ").title() # Inserção da cidade e coloca a primeira letra maiúscula
                  if cidade.replace(" ", "").isalpha(): # O replace remove os espaços vazios e o isalpha verifica se o valor digitado são letras e não está vazio
                    funcionario['cidade'] = cidade
                    break
                  print("O campo 'Cidade' não pode estar vazio!")

              elif opcao_atualizar == 3:
                while True:
                  estado = input("Digite o estado: ").upper() # Inserção do estado e conversão para maiúsculas
                  if len(estado) == 2 and estado.isalpha(): # O isalpha verifica se o valor digitado são letras, não está vazio e tem dois digitos
                    funcionario['estado'] = estado
                    break
                  print("Estado inválido! Digite exatamente duas letras.")

              elif opcao_atualizar == 4:
                while True:
                  idade = input("Digite a nova idade do funcionário(a): ")
                  if idade.isdigit() and int(idade) > 15 and int(idade) <= 75: # Verificando se o valor digitado são números e se está entre 16 e 75 anos
                    funcionario['idade'] = int(idade)
                    break
                  print("Idade inválida! Digite uma idade válida, entre 16 e 75.")

              elif opcao_atualizar == 5:
                while True:
                  escolaridade = input("Digite a atualização da escolaridade: \nF - Fundamental \nM - Médio \nS - Superior\n").lower() # Inserção da escolaridade do funcionário, coloca para minuscula a letra digitada
                  if escolaridade.isalpha() in escolaridade_map: # Verificando se a escolaridade é válida
                    funcionario['escolaridade'] = escolaridade_map[escolaridade] # Inserção da descrição da escolaridade no dicionário
                    break
                  print('Escolaridade inválida! Digite: F, M ou S.')

              elif opcao_atualizar == 6:
                while True:
                  cargo = input("Digite o novo cargo: ").title()
                  if cargo.replace(" ", "").isalpha(): # O replace remove os espaços vazios e o isalpha verifica se o valor digitado são letras e não está vazio
                    funcionario['cargo'] = cargo
                    break
                  print("O campo 'Cargo' não pode estar vazio!")

              elif opcao_atualizar == 7:
                while True:
                  salario = input("Digite o novo salário do funcionário(a): ")
                  if salario.isdigit() and int(salario) > 0: # Verificando se o valor digitado são números e maior que zero
                    funcionario['salario'] = int(salario)
                    break
                  print("Salário inválido! Digite um valor numérico e maior que zero.")

              elif opcao_atualizar == 8:
                while True:
                  email = input("Digite o novo e-mail: ")
                  if '@' in email and (email.endswith('.com') or email.endswith('.com.br')): # Validação do email se tem @, termina com .com ou .com.br
                    funcionario['email'] = email 
                    break
                  print("E-mail inválido!")

              elif opcao_atualizar == 9:
                while True:
                  dado_novo = input("Digite os novos dados do funcionário(a): chave e valor separados por espaço (ex: cpf 00000)\n")
                  if ' ' in dado_novo and dado_novo.replace(" ", "").isalnum(): # Verifica se os caracteres são digitos e letras ou apenas números ou apenas letras
                    partes = dado_novo.split(maxsplit=1) # Divide a string em duas partes
                    if len(partes) == 2:
                      chave, valor = partes # Atribui os valores das partes para as variáveis
                      if chave and valor:
                        funcionario[chave] = valor
                        print(f"Dado '{chave}' incluído com sucesso!") # Exibição do dado incluído com sucesso
                        break

                      else:
                          print("Chave e valor não podem ser vazios.")
                    else:
                      print("Entrada inválida. Certifique-se de digitar uma chave e um valor separados por espaço.")
                  else:
                    print("Entrada inválida. Certifique-se de digitar uma chave e um valor separados por espaço.")


              elif opcao_atualizar == 10:
                print("Atualização finalizada.")
                break


              print(f"Informações do funcionário(a) '{funcionario['nome']}' atualizadas com sucesso!") # Exibição do funcionário atualizado com sucesso
              print(funcionario) # Exibição do funcionário atualizado
              break
         else:
          print("Funcionário(a) não encontrado(a)! Digite um número de id válido.")
            
  elif opcao == 3:  # Opção 3: Listar todos os funcionários cadastrados
        print("\nLista de Funcionários:")
        print("==============================")

        if len(funcionarios) < 1:  # Verifica se a lista de funcionários está vazia
            print("Não há funcionários cadastrados.")
        for funcionario in funcionarios:  # Itera sobre a lista de funcionários e exibe suas informações
            for chave, valor in funcionario.items():  # Itera sobre os itens do dicionário do funcionário
                print(f"{chave.capitalize()}: {valor}")  # Exibe o nome da chave e o valor do dicionário
            print("==============================")

  elif opcao == 4:  # Opção 4: Remover um funcionário cadastrado
        print("\nRemover Funcionário(a):")
        print("==============================")

        if not funcionarios:  # Verifica se a lista de funcionários está vazia
            print("Não há funcionários cadastrados!")
        else:
            id_funcionario = input('Digite o Id do(a) funcionário(a) que deseja remover: ')
            if id_funcionario.isdigit():  # Verifica se o ID é um número inteiro
                id_funcionario = int(id_funcionario)  # Converte o ID para inteiro
                for funcionario in funcionarios:
                    if funcionario['id'] == id_funcionario:  # Verifica se o id existe na lista
                        funcionarios.remove(funcionario)  # Remove o funcionário da lista
                        print(f"Funcionário(a) {funcionario['nome']} removido(a) com sucesso!")
                        break
                    else:
                        print("Id inválido! Digite um número válido.")

  elif opcao == 5:  # Opção 5: Sair do programa
        print("Saindo... Até mais!")
        print("==============================")
        break  # Encerra o loop e sai do programa

  else:
        print("Opção inválida. Tente novamente.")  # Mensagem de erro para opções inválidas
```
